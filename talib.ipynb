```python
for w in range(0, len(company_index)):
    df = pd.read_excel(company_index[w]+".xlsx")
    inputs = {
    'open': df["open"],
    'high': df["high"],
    'low':  df["low"],
    'close': df["close"],
    'volume': df["volume"]
    }
    # uses close prices (default)
    out1 = abstract.RSI(inputs, timeperiod=6)
    out2 = abstract.MACD(inputs)[0]
    out3 = abstract.MOM(inputs, timeperiod=5)
    out4, out5, out6 = abstract.BBANDS(inputs)
    out7, out8 = abstract.STOCH(inputs)
    out9 = abstract.ADX(inputs, timeperiod=14)
    out10 = abstract.WILLR(inputs, timeperiod=14)
    out11 = abstract.BETA(inputs, timeperiod=5)
    out12 = abstract.TSF(inputs, timeperiod=14)
    out13 = abstract.ROCP(inputs, timeperiod=10)
    out14 = abstract.WMA(inputs, timeperiod=30)
    out15 = abstract.OBV(inputs)
    a =  pd.DataFrame(out1, columns=["RSI"])
    b = pd.DataFrame(out2, columns=["MACD"])
    c = pd.DataFrame(out3, columns=["MOM"])
    d = pd.DataFrame(out4, columns=["BBANDS_1"])
    e = pd.DataFrame(out5, columns=["BBANDS_2"])
    f = pd.DataFrame(out6, columns=["BBANDS_3"])
    g = pd.DataFrame(out7, columns=["KD_1"])
    h = pd.DataFrame(out8, columns=["KD_2"])
    i = pd.DataFrame(out9, columns=["ADX"])
    j = pd.DataFrame(out10, columns=["WILLR"])
    k = pd.DataFrame(out11, columns=["BETA"])
    l = pd.DataFrame(out12, columns=["TSF"])
    m = pd.DataFrame(out13, columns=["ROCP"])
    n = pd.DataFrame(out14, columns=["WMA"])
    o = pd.DataFrame(out15, columns=["OBV"])
    pd.concat([df,a , b, c, d, e, f, g, h, i, j, k, l, m, n, o], axis=1).to_excel(company_index[w]+".xlsx", \
                                                                 sheet_name = company_index[w], index=False)
    if(w%100==0):
        print(w)
```

```python
train_set = pd.read_excel("strategy_results_train.xlsx")
```

```python
import copy
```

```python
dataset = copy.deepcopy(train_set)
```

```python
temp = []
for i in range(0, dataset.shape[0]):
    if(i%1000==0):
        print(i, datetime.datetime.now())
    trade_company = dataset.iloc[i][0].split("_")[0]
    trade_date    = dataset.iloc[i][0].split("_")[1]
    df = pd.read_excel(trade_company+".xlsx")
    trade_date_datetime = datetime.date(int(trade_date[:4]), int(trade_date[4:]), 1) - datetime.timedelta(days=1)
    lower_index = 0
    for index in range(0, df.shape[0]):
        if(df["date"].iloc[index]< trade_date_datetime):
            lower_index = index
        else:
            break
    temp.append(list(df.iloc[lower_index, :]))
```

```python
pd.DataFrame(temp).to_excel("train_output.xlsx", index=False)
```
